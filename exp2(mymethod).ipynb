```
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('/content/gender_submission.csv')#importing the gender submission csv
df
```

```
test = pd.read_csv('/content/test.csv')#importing the test.csv file 
test
```

```
train = pd.read_csv('/content/train.csv')#importing the train.csv
train
```

```
train.info()#gathering the information about the
```

```
train = train.drop(['Name'],axis=1)#droping the name column as it contains the string values
test = test.drop(['Name'],axis=1)
```

```
def binary_map(x):
 return x.map({'male': 1, "female": 0})#maping the values of male and female 
```

```
train[['Sex']] = train[['Sex']].apply(binary_map)#using apply to map the values in 'sex' column in the data
test[['Sex']] = test[['Sex']].apply(binary_map)
```

```
train.info()#again verifying the information of table
```

```
train = train.drop(['Ticket','Cabin'],axis=1)#dropping the values of ticket and cabin as they are string and unique 
test = test.drop(['Ticket','Cabin'],axis=1)
```

```
train['Embarked'] = train['Embarked'].map({'S':1,'C':2,'Q':3})#mapping the values of S , C ,Q to convert it to float type
test['Embarked'] = test['Embarked'].map({'S':1,'C':2,'Q':3})
```

```
train['Age'].isnull().sum()#finding the number of null values in age column
```

```
train.fillna(train.mean() , inplace=True)#filling the null values in column with the mean values
test.fillna(test.mean(),inplace=True)
```

```
train
```

```
test
```

```
x_train = train.iloc[:,2:].values  #creating x_variable for model(independent)
y_train = train.iloc[:,1:2].values #creating y_variable for model(dependent - 'Survived')

x_test = test.iloc[:,1:].values #creating x_test value for final prediction
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler() #scalar transformation to get all values in a range
x_train = sc.fit_transform(x_train)#fitting the transformation to x_train
x_test = sc.fit_transform(x_test)#fitting the transformation to x_test
```

```
print(x_train)
```

```
np.array(y_train)
```

```
from sklearn.linear_model import LogisticRegression#buildiing the model and applying the linear regression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)#fitting the model to data

```

```
y_pred = classifier.predict(x_test)#predicting the values for x_test
```

```
y_pred #printing the values of y_prediction
```

```
y_test = df.iloc[:,1:2].values #taking the values from gender prediction for purpose of experiment
```

```
y_test #printing the values of y_test
```

```
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))#concatinating both the y_pred and y_test
print(y_pred.shape)
```

```
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
from sklearn.metrics import accuracy_score , f1_score , precision_score ,recall_score
ac = accuracy_score(y_pred,y_test) #finding the accuracy
f1 = f1_score(y_pred,y_test) #finding the f1 score
ps = precision_score(y_pred,y_test)#finding the precision score
rs = recall_score(y_pred,y_test)#finding the recall

print(ac)
print(f1)
print(ps)
print(rs)
```
