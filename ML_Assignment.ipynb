```
new = pd.DataFrame()
```

```
def tenth(n):
    if n < 50:
        return '<50'
    elif n >=50 and n < 60:
        return '50-60'
    elif n >= 60 and n < 70:
        return '60-70'
    elif n >= 70 and n < 80:
        return '70-80'
    elif n >= 80 and n < 90:
        return '80-90'
    else:
        return '90-100'
new['10th'] = df['Matric Percentage (Xth)'].apply(lambda x: tenth(x))
```

```
def twelveth(n):
    if n < 50:
        return '<50'
    elif n >=50 and n < 60:
        return '50-60'
    elif n >= 60 and n < 70:
        return '60-70'
    elif n >= 70 and n < 80:
        return '70-80'
    elif n >= 80 and n < 90:
        return '80-90'
    else:
        return '90-100'
new['12th'] = df['XII Percentage %'].apply(lambda x: twelveth(x))
```

```
def cgpa(n):
    if n < 5:
        return '<5'
    elif n >=5 and n < 6:
        return '5-6'
    elif n >= 6 and n < 7:
        return '6-7'
    elif n >= 7 and n < 7.5:
        return '7-7.5'
    elif n >= 7.5 and n < 8:
        return '7.5-8'
    elif n >=8 and n < 9:
        return '8-9'    
    else:
        return '9-10'
new['CGPA'] = df['CGPA Till 6th Sem -20-July-2019'].apply(lambda x: cgpa(x))
```

```
def LPA(n):
    if n < 3:
        return '<3'
    elif n >=3 and n < 4:
        return '3-4'
    elif n >=4 and n < 5:
        return '4-5'
    elif n >=5 and n < 6:
        return '5-6'   
    elif n >=6 and n < 10:
        return '6-10'         
    else:
        return '>10'
new['LPA'] = df['Package To (in LPA)'].apply(lambda x: LPA(x))
```

```
def GEN(n):
  if n == 'Male':
    return 'male'
  else :
    return 'female'  
new['GEN'] = df['Gender'].apply(lambda x: GEN(x))    
```
